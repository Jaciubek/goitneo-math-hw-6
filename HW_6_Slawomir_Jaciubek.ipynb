Task 1

You have ordered a study on employee work efficiency and received a functional dependency: the efficiency of employees when working on uniform tasks closely follows the formula:

y = -3x^2 + 30x


where y is the company's income generated by the developer, x is the number of working hours per day.

1. Construct the function graph and the function's derivative graph on one plot.
2. Calculate the maximum efficiency point and the efficiency value at this point analytically.
3. Construct the graph of the work efficiency function and mark the point of maximum efficiency.
4. Calculate the company benefit value that a developer brings during an 8-hour working day.



1.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# 1 step - Calculating the derivative of a function:

x = sp.symbols('x')
f = -3*x**2 + 30*x
df = sp.diff(f, x)

print("The derivative of the function f(x) = -3*x^2 + 30*x with respect to x is:", df)


def f(x):
    return -3*x**2 + 30*x

def df(x):
    return -6*x + 30

# generate x values:
x = np.linspace(0, 10, 400)

# calculate y values and y derivative for function:
y_f = f(x)
y_df = df(x)

# create the graph:
plt.figure(figsize=(8, 8))
plt.plot(x, y_f, label='y = -3x² + 30x', color='lightgreen')
plt.plot(x, y_df, label='y = -6x + 30', linestyle=':', color='magenta')
plt.title('Graph of the function and its derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# create the graph with max efficiency point:

x_max_efficiency = 5
y_max_efficiency = 75
plt.figure(figsize=(8, 8))
plt.plot(x, y_f, label='y = -3x² + 30x', color='lightgreen')
plt.scatter([x_max_efficiency], [y_max_efficiency], color='magenta')
plt.text(x_max_efficiency, y_max_efficiency - 7, f'Max({x_max_efficiency}, {y_max_efficiency})',
         horizontalalignment='center', color='magenta')
plt.title('Work efficiency and maximum efficiency point')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

4.


```python
# calculate the income for 8 hours of work using the efficiency function:

x_income = 8
y_income = f(x_income)

print("The income brought by an employee for 8 hours of work is:", y_income)
```

Task 2

You created a policy of interaction between employees, team leaders, and project managers and in result the function of employee work efficiency has changed. Now it looks like this:

y = -3(x-2)^2 + 30x


Evaluate the efficiency after changing the work scheme for employees: how much more company income does one employee bring during the 8-hour working day?

1.


```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# 1 step - Calculating the derivative of a function:

x = sp.symbols('x')
f = -3*(x-2)**2 + 30*x
df = sp.diff(f, x)

print("The derivative of the function f(x) = -3*(x-2)^2 + 30*x with respect to x is:", df)


def f(x):
    return -3*(x-2)**2 + 30*x

def df(x):
    return -6*x + 42

# generate x values:
x = np.linspace(0, 10, 400)

# calculate y values and y derivative for function:
y_f = f(x)
y_df = df(x)

# create the graph:
plt.figure(figsize=(8, 8))
plt.plot(x, y_f, label='y = -3(x-2)² + 30x', color='lightgreen')
plt.plot(x, y_df, label='y = -6x + 42', linestyle=':', color='magenta')
plt.title('Graph of the function and its derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
